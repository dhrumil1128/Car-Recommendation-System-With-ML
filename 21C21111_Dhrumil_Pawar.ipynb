```python
import pandas as pd

car_dataset = pd.read_csv(r"CARS_1.csv")
car_dataset

car_dataset['fuel_tank_capacity']=car_dataset['fuel_tank_capacity'].astype(dtype={'fuel_tank_capacity': 'string'})
car_dataset['rating']=car_dataset['rating'].astype(dtype={'rating': 'string'})
car_dataset['transmission_type']=car_dataset['transmission_type'].astype(dtype={'transmission_type': 'string'})
car_dataset['reviews_count']=car_dataset['reviews_count'].astype(dtype={'reviews_count': 'string'})
car_dataset['starting_price']=car_dataset['starting_price'].astype(dtype={'starting_price': 'string'})
car_dataset['ending_price']=car_dataset['ending_price'].astype(dtype={'ending_price': 'string'})


car_dataset.isnull().sum()
car_dataset.duplicated().sum()


#car_dataset[['Sr No.','car_name','reviews_count','fuel_type','transmission_type','rating','starting_price','ending_price']]

print("\n\n---------------------------------------<--WELCOME TO CAR RECOMMENDED SYSTEM------>------------------------\n")

def rec_rating():
    a = input("You will get a car based on how many ratings you see?\n")
    new_df=car_dataset[car_dataset['rating'] == a]
    print("\n")
    

    b = input("Which fuel type want you?\n Petrol\n Diesel\n CNG\n Electric\n")
    new_df_Fuel=new_df[new_df['fuel_type'] == b]
    print("\n")
    

    c = input("what is your starting price?\n")
    new_df_starting_price = new_df_Fuel[new_df_Fuel['starting_price'] >= c]
    print("\n")

    e = input("what is your ending price?\n")
    new_df_ending_price = new_df_starting_price[new_df_starting_price['ending_price'] <= e]
    print("\n")


    d = input("what kind transmission type do you want?\n Such as\n Manual\n Electric\n Automatic\n")

    print("See Your Car, which is recommeded according your features:--")

    print(new_df_ending_price[new_df_ending_price['transmission_type'] == d])
    print("\n")

rec_rating()

print("\n\n---------------------------------------<----THANK YOU---\U0001f600--------------->-------------------------\n")
```
